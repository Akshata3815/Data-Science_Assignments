```python
#	Calculate the mean, median, mode, and standard deviation for these columns.
#create one function of the calculate statastics
def calculate_Statistical(new_df):
    summary=pd.DataFrame(columns=['Mean','Median','Mode','Standared Deviation'])   #creating one dataframe

    for col in new_df.columns:
        mean = data[col].mean()
        median=data[col].median()
        mode=new_df[col].mode()
        std_dev = data[col].std()

        summary.loc[col] = [mean,median,mode, std_dev]

    return summary
```

```python
#calculate the statastics and return the function
satistic=calculate_Statistical(new_df)
```

```python
satistic
```

```python
# Interpretation:

### Volume :
#### The mean volume is slightly higher than the median, indicating a right-skewed distribution.
#### The standard deviation is relatively large compared to the mean, suggesting variability in the volume of sales.
#### mode suggests that a significant portion of the observations have a volume of 3

### Avg Price:
#### The mean average price is much higher than the median, suggesting a right-skewed distribution with potential outliers.
#### The standard deviation is quite large relative to the mean, indicating high variability in average prices.
#### mode suggests that 400,450,500,1300,8100 are the values most common in the dataset.

### Total Sales Value:
#### The mean total sales value is considerably higher than the median, indicating a right-skewed distribution with potential outliers.
#### The standard deviation is large relative to the mean, suggesting high variability in total sales values.
#### mode suggesting most of the values are 24300

### Discount Rate (%) :
#### The mean discount rate is slightly lower than the median, suggesting a left-skewed distribution.
#### The standard deviation is relatively small compared to the mean, indicating less variability in discount rates.

### Discount Amount :
#### The mean discount amount is higher than the median, indicating a right-skewed distribution with potential outliers.
#### The standard deviation is large relative to the mean, suggesting high variability in discount amounts.

### Net Sales Value :
#### The mean net sales value is higher than the median, indicating a right-skewed distribution with potential outliers.
#### The standard deviation is large relative to the mean, suggesting high variability in net sales values.
```

```python
#Standardization, also known as z-score normalization, is a technique used to transform numerical data in a dataset so that it has a mean of 0 and a standard deviation of 1.
#This process makes the data more interpretable and comparable across different variables by removing the units of measurement and scaling the values relative to their distribution.

#The formula for standardization, or calculating the z-score, for a numerical column ð‘¥ with mean Î¼ and standard deviation ðœŽ
#is : ð‘§=ð‘¥âˆ’ðœ‡/ðœŽ
#explanation of the process:
#Calculate Mean and Standard Deviation : Compute the mean (Î¼) and standard deviation (Ïƒ) of the numerical column.
#Subtract Mean: Subtract the mean (Î¼) from each value in the column. This centers the data around 0.
#Divide by Standard Deviation: Divide the centered values by the standard deviation (Ïƒ). This scales the data, ensuring that the standard deviation of the standardized data is 1.
#By standardizing the data, you transform it into a distribution with a mean of 0 and a standard deviation of 1, which makes it easier to interpret and compare different variables
```
